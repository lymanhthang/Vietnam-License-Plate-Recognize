```python
import os
import shutil
import random

# --- Cấu hình các đường dẫn ---
# Thư mục gốc chứa data cũ và mới
base_dir = '/content/drive/MyDrive/Do_An_Nganh/lp_detect'

# Đường dẫn đến thư mục data cũ
source_data_dir = os.path.join(base_dir, 'data', 'val')
source_images_dir = os.path.join(source_data_dir, 'images')
source_labels_dir = os.path.join(source_data_dir, 'labels')

# Đường dẫn đến thư mục data mới cho quantization
dest_data_dir = os.path.join(base_dir, 'qtz_data', 'val')
dest_images_dir = os.path.join(dest_data_dir, 'images')
dest_labels_dir = os.path.join(dest_data_dir, 'labels')

# Số lượng file cần sao chép
num_files_to_copy = 300

# --- Bắt đầu thực thi ---

def copy_files():
    """Hàm chính để thực hiện việc sao chép file."""

    print("Bắt đầu quá trình sao chép file...")

    # 1. Tạo các thư mục đích nếu chúng chưa tồn tại
    os.makedirs(dest_images_dir, exist_ok=True)
    os.makedirs(dest_labels_dir, exist_ok=True)
    print(f"Đã đảm bảo các thư mục đích tồn tại:\n- {dest_images_dir}\n- {dest_labels_dir}")

    # 2. Lấy danh sách tất cả các ảnh từ thư mục nguồn
    try:
        all_images = [f for f in os.listdir(source_images_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    except FileNotFoundError:
        print(f"LỖI: Không tìm thấy thư mục ảnh nguồn tại: {source_images_dir}")
        return

    # 3. Kiểm tra xem có đủ ảnh để sao chép không
    if len(all_images) < num_files_to_copy:
        print(f"LỖI: Thư mục nguồn chỉ có {len(all_images)} ảnh, không đủ {num_files_to_copy} ảnh để sao chép.")
        return

    # 4. Trộn ngẫu nhiên danh sách ảnh và chọn ra 100 ảnh đầu tiên
    random.shuffle(all_images)
    images_to_copy = all_images[:num_files_to_copy]
    print(f"Đã chọn ngẫu nhiên {len(images_to_copy)} ảnh để sao chép.")

    # 5. Lặp qua danh sách ảnh đã chọn và sao chép cả ảnh và file label tương ứng
    copied_count = 0
    for image_name in images_to_copy:
        # Tách tên file và phần mở rộng để tìm file label
        base_filename, _ = os.path.splitext(image_name)
        label_name = base_filename + '.txt'

        # Tạo đường dẫn đầy đủ cho file nguồn và đích
        src_image_path = os.path.join(source_images_dir, image_name)
        src_label_path = os.path.join(source_labels_dir, label_name)

        dest_image_path = os.path.join(dest_images_dir, image_name)
        dest_label_path = os.path.join(dest_labels_dir, label_name)

        # Kiểm tra sự tồn tại của cả ảnh và label trước khi sao chép
        if os.path.exists(src_image_path) and os.path.exists(src_label_path):
            shutil.copy(src_image_path, dest_image_path)
            shutil.copy(src_label_path, dest_label_path)
            copied_count += 1
        else:
            print(f"Cảnh báo: Bỏ qua '{image_name}' vì không tìm thấy file ảnh hoặc label tương ứng.")

    print(f"\nHoàn tất! Đã sao chép thành công {copied_count}/{num_files_to_copy} cặp ảnh và label.")


copy_files()
```

```python
import os

def check_data_consistency(data_directory):
    """
    Kiểm tra sự khớp nhau giữa các file trong thư mục 'images' và 'labels'.

    Args:
        data_directory (str): Đường dẫn đến thư mục cần kiểm tra (ví dụ: 'data/val').
    """
    images_dir = os.path.join(data_directory, 'images')
    labels_dir = os.path.join(data_directory, 'labels')

    print(f"--- Bắt đầu kiểm tra thư mục '{data_directory}' ---")

    # 1. Kiểm tra sự tồn tại của các thư mục
    if not os.path.isdir(images_dir) or not os.path.isdir(labels_dir):
        print(f"LỖI: Không tìm thấy thư mục 'images' hoặc 'labels' bên trong '{data_directory}'.")
        print("Vui lòng kiểm tra lại cấu trúc thư mục.")
        return

    # 2. Lấy danh sách tên file (không có phần mở rộng) và đưa vào Set
    # Set là cấu trúc dữ liệu hiệu quả để so sánh và tìm khác biệt
    image_basenames = {os.path.splitext(f)[0] for f in os.listdir(images_dir)}
    label_basenames = {os.path.splitext(f)[0] for f in os.listdir(labels_dir)}

    print(f"Tìm thấy tổng cộng: {len(image_basenames)} ảnh và {len(label_basenames)} label.")

    # 3. Tìm sự khác biệt giữa hai tập hợp tên file
    images_missing_labels = image_basenames - label_basenames
    labels_missing_images = label_basenames - image_basenames

    # 4. In kết quả
    has_mismatch = False
    if images_missing_labels:
        has_mismatch = True
        print(f"\n VẤN ĐỀ: Có {len(images_missing_labels)} ảnh bị THIẾU file label:")
        # Chỉ in ra tối đa 10 ví dụ để tránh làm rối màn hình
        for name in sorted(list(images_missing_labels))[:10]:
            print(f"  - {name}")
        if len(images_missing_labels) > 10:
            print("  - ... và một số file khác.")

    if labels_missing_images:
        has_mismatch = True
        print(f"\n VẤN ĐỀ: Có {len(labels_missing_images)} file label bị THIẾU ảnh:")
        for name in sorted(list(labels_missing_images))[:10]:
            print(f"  - {name}")
        if len(labels_missing_images) > 10:
            print("  - ... và một số file khác.")

    if not has_mismatch:
        print("\n TUYỆT VỜI! Mọi thứ đều khớp nhau hoàn hảo.")

    print("\n--- Kết thúc kiểm tra ---")


if __name__ == "__main__":

    TARGET_DIRECTORY = '/content/drive/MyDrive/Do_An_Nganh/lp_detect/qtz_data/val'

    check_data_consistency(TARGET_DIRECTORY)
```

Dự doán với model sau khi lượng tử hóa

```python
# !pip install tflite-runtime
import cv2
import numpy as np
import tflite_runtime.interpreter as tflite
from google.colab.patches import cv2_imshow

# --- CÁC THAM SỐ CẦN THAY ĐỔI ---
MODEL_PATH = "/content/drive/MyDrive/Do_An_Nganh/lp_detect/runs/detect/lp_t4_train/weights/best_saved_model/best_int8.tflite"  
IMAGE_PATH = "/content/drive/MyDrive/Do_An_Nganh/lp_detect/qtz_data/val/images/Dieu_0017.png"      
CLASSES = ['License_Plate']                   
CONFIDENCE_THRESHOLD = 0.25                      
NMS_THRESHOLD = 0.4                             

def draw_detections(image, boxes, scores, class_ids):
    """Vẽ các bounding box và nhãn lên ảnh."""
    for i in range(len(boxes)):
        x1, y1, x2, y2 = map(int, boxes[i])
        confidence = scores[i]
        class_id = class_ids[i]

        label = f"{CLASSES[class_id]}: {confidence:.2f}"
        color = (0, 255, 0) # Màu xanh lá

        # Vẽ bounding box
        cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)

        # Vẽ nhãn
        (text_width, text_height), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 2)
        cv2.rectangle(image, (x1, y1 - text_height - 5), (x1 + text_width, y1), color, -1)
        cv2.putText(image, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1)

    return image

def detect(model_path, image_path):
    """Thực hiện nhận diện đối tượng trên một ảnh bằng model TFLite."""

    interpreter = tflite.Interpreter(model_path=model_path)
    interpreter.allocate_tensors()

    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()

    _, input_height, input_width, _ = input_details[0]['shape']

    original_image = cv2.imread(image_path)
    original_h, original_w, _ = original_image.shape

    resized_image = cv2.resize(original_image, (input_width, input_height))
    input_data = np.expand_dims(resized_image, axis=0)

    if input_details[0]['dtype'] == np.float32:
        input_data = (input_data / 255.0).astype(np.float32)

    interpreter.set_tensor(input_details[0]['index'], input_data)
    interpreter.invoke()

    output_data = interpreter.get_tensor(output_details[0]['index'])
    detections = np.squeeze(output_data).T

    boxes, scores, class_ids = [], [], []

    for det in detections:
        class_scores = det[4:]
        confidence = np.max(class_scores)
        class_id = np.argmax(class_scores)

        if confidence > CONFIDENCE_THRESHOLD:
            cx, cy, w, h = det[:4]

            x1 = int((cx - w / 2) * original_w)
            y1 = int((cy - h / 2) * original_h)
            x2 = int((cx + w / 2) * original_w)
            y2 = int((cy + h / 2) * original_h)

            boxes.append([x1, y1, x2 - x1, y2 - y1]) 
            scores.append(float(confidence))
            class_ids.append(class_id)

    indices = cv2.dnn.NMSBoxes(boxes, scores, CONFIDENCE_THRESHOLD, NMS_THRESHOLD)

    final_boxes, final_scores, final_class_ids = [], [], []
    if len(indices) > 0:
        for i in indices.flatten():
            x, y, w, h = boxes[i]
            final_boxes.append([x, y, x + w, y + h]) 
            final_scores.append(scores[i])
            final_class_ids.append(class_ids[i])

    result_image = draw_detections(original_image.copy(), final_boxes, final_scores, final_class_ids)

# 8. Hiển thị và lưu ảnh
    print("Đang hiển thị ảnh kết quả...")
    from google.colab.patches import cv2_imshow
    cv2_imshow(result_image)

    cv2.imwrite("result.jpg", result_image)
    print("\nĐã lưu ảnh kết quả vào file result.jpg")


if __name__ == "__main__":
    detect(MODEL_PATH, IMAGE_PATH)
```
